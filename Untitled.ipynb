### Monte Carlo Process to estimate ùúã with visualization

The task is to compute an approximation to $ \pi $ using [Monte Carlo](https://en.wikipedia.org/wiki/Monte_Carlo_method).


## Version 1
Draw 
$$
x^2 + y^2 = 1
$$ 
to the coordinate and plot the dots.

```python
import numpy as np
import matplotlib.pyplot as plt

# Initial Setting
# trial is the number of trial
# inCircle is the number of points in the circle
trial = 1000
inCircle = 0

# Draw circle with radius 1 : x^2 + y^2 = 1
circle = plt.Circle((0, 0), radius = 1, color='black', fill=False)
ax = plt.gca()
ax.add_artist(circle)

# Plot with uniform dist
for i in range(trial):
    x = np.random.uniform(-1,1)
    y = np.random.uniform(-1,1)
    
    if (x**2 + y**2 < 1):
        inCircle += 1
        plt.scatter(x, y, s = 1, color = "black")
    else:
        plt.scatter(x, y, s = 1, color = "blue")
            
# Show the canvas and calculate the answer
plt.axis([-1, 1, -1, 1])
plt.show()

ratio = inCircle/trial
œÄ = ratio * 4
print(œÄ)
```

## Version 2
Plot approximation process

```python
import numpy as np
import matplotlib.pyplot as plt
import math

def monteCarloPi(trial):
    cnt = 0
    
    for i in range(trial):
        x = np.random.uniform(-1,1)
        y = np.random.uniform(-1,1)
        
        if (x**2 + y**2 < 1):
           cnt = cnt+1
    
    return (cnt/trial)*4

def plotApproxProcess(x, œµ, trial):
    plt.hlines(œÄ, 1, trial, colors='black', linestyles='solid', label='œÄ')
    for i in range(1, trial+1):
        result = monteCarloPi(i)
        if abs(result - x) < œµ:
            plt.scatter(i, result, s = 1, color = "blue")
        else:
            plt.scatter(i, result, s = 1, color = "red")
    
œÄ = math.pi
œµ = 0.1
plotApproxProcess(œÄ, œµ, 1000)
plt.show()
```
